```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
#Seaborn is a statistical plotting library
#it has statistical graphs like distribution plots , heatmap etc.
#It has variety of graph styles
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe(include='all')#gives descriptive statistical summary for all numeroical values
```

# Distribution Plot

```python
#Dist plot is used to show the distribution of univariable set of observations(ine variable)

sns.distplot(tips['total_bill'])
#here the data is rightly skewed
```

```python
sns.distplot(tips['total_bill'], kde=False) #Removing the kernel density line
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
#most customers are paying the bill in range of 10 to 20
#most frequent bill amount - 10 to 20
```

```python
sns.distplot(tips['tip'],kde=False,bins=10)
#most of tip lie from 2 to 4 $
```

```python
#distribution plot does not work on categorical data
#sns.distplot(tips['sex'])----this will give error
```

# JOINT PLOT

```python
#A joint plot is used to show the distribution of bivariate set of observations(two variables)
#It is basically two Dist Plot
#It also shows the scatter plot between the two variables
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')#hex-hexagon representation
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')#reg-regression with best fit line
```

# PAIR PLOT

```python
#It plots pairwise relationship across an entire data frame at least for
#the numerical columns
```

```python
sns.pairplot(tips)
#we have 3 numerical variable columns so we get a 3X3 pair plot
```

```python
sns.pairplot(tips, hue='sex')#add hue(color) for categorical variable
#orange color depicts female
#blue color depicts male
```

```python
sns.pairplot(tips, hue='day')
```

# Plotting Categorical Data

## BARPLOT

```python
#It is just a general plot that allows to aggregatethe categorical 
#data based off of some function and by default thats the mean.

sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

## COUNTPLOT

```python
#Countplot is a kind of barplot in which the estimator is the count
sns.countplot(x='sex',data=tips)
```

## BOXPLOT

```python
#These are used to show the distribution of categorical data
#and it shows the distribution of quantitative data in a way that hopefully 
#facilitates comparisons between the variables

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

## VIOLIN PLOT

```python
#These are used to show the distribution of categorical data
#and it shows the distribution of quantitative data in a way that hopefully 
#facilitates comparisons between the variables
#It shows us a little more info that the box plot in the sense that it also indicates the distribution of the variable

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)
```
