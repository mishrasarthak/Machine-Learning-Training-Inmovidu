```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib Inline #so that all charts are showed in this notebook
```

```python
x= np.arange(0,11)
y=x**3
```

# Line Charts

```python
plt.plot(x,y)#on x-axis - x and y axis - y
```

```python
 #Adding axis labels and chart title
plt.plot(x,y)
plt.title('My First Chart')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

# Multiple Line Charts

```python
x= np.arange(0,11)
y=x**3

x2= np.arange(0,11)
y2=(x**3)/2

plt.plot(x,y,linewidth=5)
plt.plot(x2,y2,linewidth=2)
plt.title('Second Chart')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
```

```python
#Adding colors/legends to chart
plt.plot(x,y,'r',label='first data',linewidth=5)
plt.plot(x2,y2,'y',label='second data',linewidth=5)
plt.legend()
```

```python
#we can change the loaction of the legend
plt.plot(x,y,'r',label='first data',linewidth=5)
plt.plot(x2,y2,'y',label='second data',linewidth=5)
plt.legend(loc=10)
```

```python
plt.figure(figsize=(10,5)) #To adjust size of the chart
years = [1950,1955,1960,1965,1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2020]
pops = np.array([1262,1650,2525,2785,3222,2682,4861,4440,4853,5310,5375,6520,7430,7800,8900])
pops_b = pops/1000
plt.plot(years,pops_b,linewidth=2)
plt.ylabel('Population in billions')
plt.xlabel('Population growth by year')
plt.title('Population Growth')
plt.grid(True)#To add grid
```

```python
plt.figure(figsize=(10,5)) #To adjust size of the chart
years = [1950,1955,1960,1965,1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2020]
pops = np.array([1262,1650,2525,2785,3222,2682,4861,4440,4853,5310,5375,6520,7430,7800,8900])
pops_b = pops/1000
deaths=[1,1,0.8,1,2,2.1,3,4.2,5,5,6.3,7.6,8.9,9, 9.2]
plt.plot(years,pops_b,linewidth=2,label='Population trends',marker='*',markersize=10)
plt.plot(years,deaths,label='Mortality trends',marker='o')
plt.ylabel('Population in billions')
plt.xlabel('Population growth by year')
plt.title('Population Growth')
plt.legend()
plt.grid(True)#To add grid
```

# BAR CHARTS

```python
x=['India','USA','Japan','Russia','Australia']
y=[6,7,8,9,5]
plt.bar(x,y,label='Bars1',color='y')
plt.xlabel('Country')
plt.ylabel('Inflation rate%')
plt.title("BAR GRAPH")
```

# Multiple Bars

```python
x=['India','USA','Japan','Russia','Australia']
y=[6,7,8,9,5]
x2=['India','USA','Japan','Russia','Australia']
y2=[5,2,4,3,10]
plt.bar(x,y,label='Inflation',color='y')
plt.bar(x2,y2,label='Growth',color='r')
plt.xlabel('Country')
plt.ylabel('Inflation rate%')
plt.title("BAR GRAPH")
```

# SCATTER PLOTS

```python
#scatter plots show how much one variable is affected by another.
#The relationship between two variables is called correlation.
x=np.array([1,2,3,4,5,6,7,8])
y=x*2
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("scatter plot no.1")
```

# Pie Charts

```python
raw_data = {'names':['Nick','Sani','John','Rubi','Maya'],'jan_score':[123,123,125,126,128],
           'feb_score':[23,24,25,26,29],'march_score':[3,5,6,7,9]}
df = pd.DataFrame(raw_data,columns=['names','jan_score','feb_score','march_score'])
df['total_score']=df['jan_score']+df['feb_score']+df['march_score']
df
```

```python
plt.pie(df['total_score'],labels=df['names'],autopct='%.2f%%')
plt.axis('equal')
plt.axis('equal')
plt.show()
```

```python
plt.pie(df['total_score'],labels=df['names'],autopct='%.2f%%',explode=(0.2,0,0,0,0))
plt.axis('equal')
plt.axis('equal')
plt.show()
```

# SUBPLOTS

```python
#Horizontal Subplots
plt.figure(figsize=(15,5))
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([5,2,4,2,1,4,5,2])

plt.subplot(1,2,1)#1 row, 2 columns , 1st chart
plt.plot(x,y,'g')#plotting x on x axis and y on y axis with color as green

plt.subplot(1,2,2)#1 row, 2 columns , 2nd chart
plt.plot(y,x,'b')#plotting y on x axis and x on y axis with color as blue
```

```python
#Horizontal Subplots
plt.figure(figsize=(15,5))
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([5,2,4,2,1,4,5,2])

plt.subplot(1,3,1)#1 row, 3 columns , 1st chart
plt.plot(x,y,'g')#plotting x on x axis and y on y axis with color as green
plt.title('Graph-1')

plt.subplot(1,3,2)#1 row, 3 columns , 2nd chart
plt.plot(y,x,'b')#plotting y on x axis and x on y axis with color as blue
plt.title('Graph-2')

plt.subplot(1,3,3) 
plt.plot(x**2,y,'y')
plt.title('Graph-3')

#plt.tight_layout(w_pad=10)#to increase or decrease distance between the charts
```

# VERTICAL SUBPLOTS

```python
plt.figure(figsize=(15,5))
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([5,2,4,2,1,4,5,2])

plt.subplot(2,1,1)#2 rows and 1 columns for 1st chart
plt.plot(x,y,'g')

plt.subplot(2,1,2)#2 rows and 1 columns for 2nd chart
plt.plot(y,x,'b')

plt.tight_layout(h_pad=10,w_pad=5)
```

```python
#SUBPLOT 2X2


plt.subplot(2,2,1)#2 rows and 1 columns for 1st chart
plt.plot(x,y,'g')
plt.title('Chart-1')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.subplot(2,2,2)#2 rows and 1 columns for 2nd chart
plt.plot(y,x,'b')
plt.title('Chart-2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.subplot(2,2,3)#2 rows and 1 columns for 1st chart
plt.plot(y*2,x*2,'y')
plt.title('Chart-3')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.subplot(2,2,4)#2 rows and 1 columns for 2nd chart
plt.plot(x*2,y*2,'r')
plt.title('Chart-4')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.tight_layout(h_pad=5,w_pad=5)
```
