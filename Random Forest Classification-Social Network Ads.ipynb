# Random Forest Classification
Random Forest algorithm is a supervised algorithm. As the name suggest , this algorithm creates the forest with a number of trees.

topmost application of Random forest is the motion gaming.

# Working of Random Forest Algorithm

1. Assume number of cases in the training set is N, Then the sample of N cases is taken at random but with replacement. This sample will be the training set for growing the tree.

2. If there are M input variables, a number m < M is specified such that at each node, m variables are selected at random out of the M. The best split on these m is used to split the node. The value m is held constant while we grow the forest, three possible values for m : 1/2M, sqrt(M) , 2*sqrt(M)

3. Each tree is grown to the largest extent possible. Predict new data by aggregating the prediction of the ntree trees(i.e majority votes for classification, average for regression.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://s3.amazonaws.com/thinkific/file_uploads/287654/attachments/2e3/a35/b45/Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop('User ID',axis=1,inplace=True)
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
cols = ['Age','EstimatedSalary','Gender_Male','Purchased']
df= df[cols]
df
```

```python
X = df.iloc[:,:-1].values
X.shape
```

```python
Y = df.iloc[:,-1].values
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train =sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 10, n_jobs =2 , random_state=0)
#n_estimators = number of decision trees, n_jobs = number of computers.
clf.fit(X_train , Y_train)
```

```python
Y_pred = clf.predict(X_test)
Y_pred
```

```python
Y_test
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_test , Y_pred)
```

```python
cm
```

```python
metrics.accuracy_score(Y_test, Y_pred)
#ACCURACY
```
